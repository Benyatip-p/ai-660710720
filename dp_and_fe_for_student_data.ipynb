คำอธิบาย: จัดการ missed value, ค่าที่ผิดปกติ ใน filed "อายุ" โดยใช้ค่าเฉลี่ยอายุจากข้อมูลที่ปกติในแต่ละแถวมาใส่ในค่าว่างแทน และการจัดการข้อมูลที่ไม่ปกติจะทำการตรวจสอบอายุที่เกิน 100 ปีและแทนด้วย NaN จากนั้นแทนด้วยค่าเฉลี่ย

เหตุผล เพราะจะได้ไม่ต้องลบข้อมูลในแถวใดแถวนึงทิ้งและยังเลือกใช้ข้อมูลที่อยู่ในช่วงที่ปกติมาอนุมานข้อมูลอื่นได้

```python
valid_ages = df.loc[(df['อายุ'].notna()) & (df['อายุ'] <= 100), 'อายุ']
mean_age = valid_ages.mean().round(0)

df.loc[df['อายุ'] > 100, 'อายุ'] = mean_age

df['อายุ'] = df['อายุ'].fillna(mean_age)
df
```

คำอธิบาย: จัดการข้อมูลที่แตกต่างกันของฟิลด์คณะโดยการใช้ฟังชั่น ในการเทียบชื่อคณะในรูปแบบที่แตกต่างกันแต่ละแถวและให้ return ออกมาเป็นชื่อที่เป็นมาตรฐาน + คลีนข้อมูลคณะที่เป็น missed โดยการใช้คำว่า 'ไม่ระบุ' แทน

เหตุผล: ข้อมูลคณะในแถวมีมากกว่า 2 เลยไม่เลือกใช้การเดาดีกว่า

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวะ', 'วิศวฯ', 'Engineering', 'วิศวกรรม']:
        return 'วิศวกรรมศาสตร์'
    else:
        return 'ไม่ระบุ'
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
df
```

คำอธิบาย: จัดการข้อมูลเหมือนกันการจัดการข้อมูลใน field คณะ
แต่เนื่องจากข้อมูลของเพศมีความจำเพาะกว่าจึงไม่สามารถใช้การเฉลี่ยหรือเดาตามจำนวนเพศที่ลงทะเบียนมากที่สุดแล้วนำมาแทนไม่ได้ (ความเห็นส่วนตัว)

```python
def standardize_gender(name):
    if name in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif name in ['หญิง', 'Female', 'ญ', 'F']:
        return 'หญิง'
    else:
        return 'ไม่ระบุ'
```

```python
df['เพศ'] = df['เพศ'].apply(standardize_gender)
df
```

คำอธิบาย: ใช้ค่าเฉลี่ยอนุมานน้ำหนักของค่าว่างภายในตารางเพราะเป็นข้อมูลตัวเลขและอิงจากข้อมูลในตารางทั้งหมด

```python
print(df['น้ำหนัก_kg'].dtype)
df['น้ำหนัก_kg'] = pd.to_numeric(df['น้ำหนัก_kg'], errors='coerce')
```

```python
mean_weight = df['น้ำหนัก_kg'].mean()
```

```python
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].fillna(mean_weight).round(1)
```

```python
df
```

คำอธิบาย: ใช้การประมาณค่าเฉลี่ยแทนค่าส่วนสูงที่เป็นค่าว่าง เพราะเป็นข้อมูลประเภทตัวเลขแล้วก็อิงจากข้อมูลส่วนมากในตาราง

```python
mean_height = df['ส่วนสูง_cm'].mean()
```

```python
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(mean_height).round(1)
```

```python
df
```

คำอธิบาย: ใช้ค่า mean คำนวณและแทนค่าในค่าที่ ชั่วโมงเรียนเป็นค่าว่างเพราะเป็นข้อมูลแบบตัวเลขและสามารถหาค่าเฉลี่ยได้

```python
mean_hr = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()
df['ชั่วโมงเรียน_ต่อสัปดาห์'] = df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hr).round(1)
```

```python
df
```

คำอธิบาย:  ทำการประมาณค่า GPA โดยใช้ mean เนื่องจากมีทั้งค่าที่เป็น NaN และ ค่าที่ผิดปกติคือ > 4.00 และ < 1.00

เหตุผล ที่ใช้เพราะข้อมูลเป็นตัวเลข

```python
valid_GPA = df.loc[(df['GPA'].notna()) & (df['GPA'] >= 1) & (df['GPA'] <= 4), 'GPA']
mean_GPA = valid_GPA.mean().round(1)

df.loc[df['GPA'] > 4, 'GPA'] = mean_GPA
df.loc[df['GPA'] < 1, 'GPA'] = mean_GPA

df['GPA'] = df['GPA'].fillna(mean_GPA)

df
```

คำอธิบาย: ทำการคลีนชื่อ โดยการตัดแถวที่ชื่อเป็นค่าว่างทิ้งเพราะพิจารณาจากข้อมูลในตารางแล้วมีเพียงแค่แถวเดียวที่ไม่มีชื่อ

```python
df = df.dropna(subset=['ชื่อ'])
df
```
